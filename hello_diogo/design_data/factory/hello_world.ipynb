```python
from pathlib import Path
import gdsfactory as gf
```

```python
from ihp import PDK
PDK.activate()
```

```python
list(PDK.layers)
```

```python
GDS_PATH = Path('../gds').resolve()
GDS_PATH
```

```python

# Create a new component
c = gf.Component()

# Add text elements
t1 = gf.components.text("Hello", size=10, layer=PDK.layers.TopMetal1drawing)
t2 = gf.components.text("Diogo", size=10, layer=PDK.layers.TopMetal1drawing)

text1 = c.add_ref(t1)
text2 = c.add_ref(t2)

# Position elements
text2.xmin = text1.xmax + 2


# Show the result
c.show()
```

```python
c.write(GDS_PATH /  "hello_diogo.gds")
```
